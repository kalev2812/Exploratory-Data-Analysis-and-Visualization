```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
 
df=pd.read_csv('/content/Placement_Data_Full_Class.csv')
df.head()
```

```python
sns.lineplot(x='ssc_p',y='hsc_p',data=df)
```

```python
sns.factorplot(x='workex', data=df, kind='count')
```

```python
sns.factorplot(x='workex', data=df, kind='count', hue='gender')
```

```python
df['salary'].hist(bins=8)
```

```python
as_fig = sns.FacetGrid(df,hue='gender',aspect=5)
as_fig.map(sns.kdeplot,'salary',shade=True)
oldest = df['salary'].max()
as_fig.set(xlim=(0,oldest))
as_fig.add_legend()
```

```python
df.plot.scatter(x='ssc_p',y='hsc_p')
```

```python
df.plot.scatter(x='degree_p',y='degree_t')
```

```python
import matplotlib.pyplot as plt
sizes= df['workex'].value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(sizes,labels=['Not workex', 'workex'],autopct='%1.1f%%',shadow=True)
plt.show()
```
